### Langkah-Langkah Preprocessing Data
### 1. Data Cleansing (Remove Regex)
Tujuan dari Cleansing ini adalah untuk Menghapus tanda baca, url, hashtagh,karakter, angka dan lain-lain

```python
def remove(Text):
    # remove angka
    Text = re.sub('[0-9]+', '', Text)
    # Strip space, " and ' from text
    Text = Text.strip(' "\'')
    # replace punctuation with space
    Text = re.sub(f"[{re.escape(string.punctuation)}]", " ", Text)
    # Replace multiple spaces with a single space
    Text = re.sub(r'\s+', ' ', Text)
    
    return Text

data['clean'] = data['benefit'].apply(lambda x: remove(x))

data.head()
```

### 2. Case Folding
Tujuan dari case folding untuk mengubah semua huruf dalam dokumen menjadi huruf kecil. Hanya huruf ‘a’ sampai ‘z’ yang diterima.

```python
def clean(text):
  # Mengubah teks ke lowercase
  text = text.lower()


  # Menghapus non ASCII character
  encoded_string = text.encode("ascii", "ignore")
  text = encoded_string.decode()

  return(text)
```

```python
data['clean'] = data['clean'].astype(str)
data['Case_Folding'] = data['clean'].apply(lambda x:clean(x))
data
```

### 4. Tokenizing
Tokenizing adalah proses pemisahan teks menjadi potongan-potongan yang disebut sebagai token

```python
def word_tokenize_wrapper(text):
 return word_tokenize(text)
data['Tokenizing'] = data['Case_Folding'].apply(word_tokenize_wrapper)
data
```

### 4. Filtering (Stopword Removal)
Filtering adalah tahap mengambil kata-kata penting dari hasil token dengan menggunakan algoritma stoplist (membuang kata kurang penting) atau wordlist (menyimpan kata penting).

```python
#Proses Stopword Removal
def stopword_removal(text):
    filtering = stopwords.words('indonesian','english')
    x = []
    data = []
    def myFunc(x):
        if x in filtering:
            return False
        else:
            return True
    fit = filter(myFunc, text)
    for x in fit:
        data.append(x)
    return data
data['Filtering'] = data['Tokenizing'].apply(stopword_removal)
data
```

### 5. Stemming
Stemming adalah proses menghilangkan infleksi kata ke bentuk dasarnya, namun bentuk dasar tersebut tidak berarti sama dengan akar kata (root word). Misalnya kata “mendengarkan”, “dengarkan”, “didengarkan” akan ditransformasi menjadi kata “dengar

```python
# proses stemming
def stemming(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in text:
        dt = stemmer.stem(w)
        do.append(dt)
    d_clean=[]
    d_clean=" ".join(do)
    print(d_clean)
    return d_clean
data['Stemming'] = data['Filtering'].apply(stemming)
data
```

### Proses Lemmatisasi untuk teks bahasa inggris

```python
nltk.download('averaged_perceptron_tagger_eng')

# Membuat objek WordNetLemmatizer untuk proses lemmatization
lemmatizer = WordNetLemmatizer()

# Membuat mapping POS tag ke tipe yang sesuai di WordNet (NOUN, VERB, ADJ, ADV)
wordnet_map = {"N": wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}

# Fungsi untuk melakukan lemmatization pada teks
def lemmatize_words(text):
    # POS tagging tanpa lang
    pos_tagged_text = nltk.pos_tag(text.split())

    return " ".join([
        lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN))
        for word, pos in pos_tagged_text
    ])

# Terapkan ke kolom
data["Lemmatization"] = data["Stemming"].apply(lambda text: lemmatize_words(text))

# Menampilkan data dengan kolom baru 'Lemmatization'
data
```

```python
# Load data revisi kata

revword_data = pd.read_excel("revword.xlsx")

# Membuat kamus pengganti dari kolom 'Kata' dan 'Revisi'
replacement_dict = dict(zip(revword_data['Kata'], revword_data['Revisi']))

# Fungsi untuk mengganti kata berdasarkan kamus pengganti
def ganti_kata(teks):
    kata_kata = teks.split()  # Memisahkan teks menjadi kata-kata
    teks_revisi = [replacement_dict.get(kata, kata) for kata in kata_kata]  # Mengganti kata jika ada di kamus
    return ' '.join(teks_revisi)  # Menggabungkan kembali kata-kata yang sudah diganti

# Terapkan fungsi penggantian pada kolom yang berisi teks (misalnya kolom 'Teks')
data['Benefit_Clean'] = data['Lemmatization'].apply(ganti_kata)
# Tampilkan data
data
```

```python
# Daftar kata yang ingin dibuang
stopwords = ["nggak", "yg", "men", "me", "mengemba", "mu", "nya"]

# Fungsi untuk menghapus stopwords
def remove_stopwords(text):
    # Menggunakan regular expressions untuk mengganti kata-kata yang ada dalam daftar stopwords
    pattern = r'\b(?:' + '|'.join(stopwords) + r')\b'
    return re.sub(pattern, '', text)

# Terapkan fungsi ke kolom 'text' di DataFrame
data['Benefit_Clean'] = data['Benefit_Clean'].apply(remove_stopwords)
```
