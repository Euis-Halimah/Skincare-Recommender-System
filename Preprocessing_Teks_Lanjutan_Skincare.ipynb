```python
# Import Library
import pandas as pd
import re
import string
```

```python
# Import Data
data = pd.read_csv('hasil preprocessing benefit.csv')
data
```

### Preprocessing untuk kolom ingredients

```python
# Fungsi untuk pembersihan teks
def clean_ingredients(ingredient):
    # Mengubah teks menjadi huruf kecil
    ingredient = ingredient.lower()
    # Hapus karakter kontrol ASCII (non-printable: \x00 - \x1F dan \x7F)
    ingredient = re.sub(r'[\x00-\x1F\x7F]', '', ingredient)
    # Hapus karakter non-ASCII (misal emoji, huruf asing)
    ingredient = re.sub(r'[^\x00-\x7F]', '', ingredient)
    # Strip space, " and ' from text
    ingredient = ingredient.strip(' "\'')
    # remove angka
    ingredient = re.sub('[0-9]+', '', ingredient)
    # Remove punctuation (hapus tanpa ganti spasi)
    ingredient = re.sub(f"[{re.escape(string.punctuation)}]", " ", ingredient)
    # Replace multiple spaces with a single space
    ingredient = re.sub(r'\s+', ' ', ingredient)
    # Memisahkan bahan berdasarkan koma yang setelahnya ada spasi
    #ingredient_list = ingredient.split()
    return ingredient

# Terapkan pembersihan pada kolom 'ingredient'
data['cleaned_ingredient'] = data['ingredients'].apply(clean_ingredients)

data
```
