```python
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

# Misal kamu sudah punya tfidf_matrix (dalam bentuk numpy array atau csr_matrix)
# Pilih produk yang dijadikan acuan, misalnya index ke-8
product_index = 1023

# Hitung Euclidean distance antara produk tersebut dengan semua produk
distances = euclidean_distances(tfidf_combined[product_index], tfidf_combined).flatten()

# Urutkan dari yang paling mirip (jarak paling kecil), kecuali dirinya sendiri (jarak = 0)
similar_indices = distances.argsort()[1:6]  # ambil 5 terdekat selain dirinya sendiri

# Ambil jarak-jarak ke produk acuan
similar_distances = distances[similar_indices]

# Buat DataFrame rekomendasi
rekomendasi_df = data.loc[similar_indices, ['name', 'brand', 'sub-category', 'cleaned_ingredient', 'Benefit_Clean']].copy()
rekomendasi_df['Euclidean_distance'] = similar_distances

# Tampilkan DataFrame rekomendasi
rekomendasi_df
```

```python
def get_recommendations(input_benefit, input_ingredients, input_subcat, jumlah):
    import pandas as pd, re, string
    from nltk.tokenize import word_tokenize
    from nltk.corpus import stopwords, wordnet
    from nltk.stem import WordNetLemmatizer
    from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.metrics.pairwise import euclidean_distances
    from scipy.sparse import hstack
    import nltk

    nltk.download('punkt')
    nltk.download('stopwords')
    nltk.download('wordnet')
    nltk.download('averaged_perceptron_tagger_eng')

    wordnet_map = {"N": wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}

    # Load data
    df = pd.read_excel("D:\Documents\Tugas Akhir\Recommender System\data_cleaned.xlsx")
    rev_data = pd.read_excel("D:/Documents/Tugas Akhir/Recommender System/revword.xlsx")
    rev_dict = dict(zip(rev_data['Kata'].astype(str), rev_data['Revisi'].astype(str)))

    def preprocess_benefit(text):
        if pd.isnull(text): return ""  # Ensure empty strings for NaN
        try:
            text = re.sub(r'\d+', '', text)
            text = text.strip(' "\'')
            text = re.sub(f"[{re.escape(string.punctuation)}]", " ", text)
            text = re.sub(r"\s+", " ", text).lower()
            text = text.encode("ascii", "ignore").decode()
            tokens = word_tokenize(text)
            stop_words = set(stopwords.words('english') + stopwords.words('indonesian'))
            tokens = [t for t in tokens if t not in stop_words and len(t) > 1]
            stemmer = StemmerFactory().create_stemmer()
            tokens = [stemmer.stem(t) for t in tokens]
            lemmatizer = WordNetLemmatizer()
            pos_tagged = nltk.pos_tag(tokens)
            tokens = [lemmatizer.lemmatize(w, wordnet_map.get(p[0], wordnet.NOUN)) for w, p in pos_tagged]
            hasil = " ".join([rev_dict.get(t, t) for t in tokens])
            return hasil if isinstance(hasil, str) else ""
        except:
            return ""

    def preprocess_ingredients(text):
        if pd.isnull(text): return ""  # Ensure empty strings for NaN
        text = text.lower()
        text = re.sub(r'[\x00-\x1F\x7F]', '', text)
        text = re.sub(r'[^\x00-\x7F]', '', text)
        text = text.strip(' "\'')
        text = re.sub(r'\d+', '', text)
        text = re.sub(f"[{re.escape(string.punctuation)}]", " ", text)
        text = re.sub(r'\s+', ' ', text)
        return text

    # Preprocessing input
    Benefit_Clean = preprocess_benefit(input_benefit)
    cleaned_ingredient = preprocess_ingredients(input_ingredients)
    sub_category = input_subcat.lower()

    # Adding user input to the dataframe
    df = pd.concat([df, pd.DataFrame([{
        'produk': 'Inputan User',
        'Benefit_Clean': Benefit_Clean,
        'cleaned_ingredient': cleaned_ingredient,
        'sub-category': sub_category
    }])], ignore_index=True)


    # Fill NaN values with empty string
    df[['Benefit_Clean', 'cleaned_ingredient', 'sub-category']] = df[['Benefit_Clean', 'cleaned_ingredient', 'sub-category']].fillna("")

    # TF-IDF Vectorization
    tfidf_benefit = TfidfVectorizer(ngram_range=(1,2))
    tfidf_ing = TfidfVectorizer(ngram_range=(1,3))
    tfidf_subcat = TfidfVectorizer(ngram_range=(1,3))

    tfidf_matrix_benefit = tfidf_benefit.fit_transform(df['Benefit_Clean'])
    tfidf_matrix_ing = tfidf_ing.fit_transform(df['cleaned_ingredient'])
    tfidf_matrix_subcat = tfidf_subcat.fit_transform(df['sub-category'])


    # Combine the matrices
    combined_matrix = hstack([tfidf_matrix_benefit * 0.25, tfidf_matrix_ing * 0.25, tfidf_matrix_subcat * 0.5])

    # Calculate Euclidean distance
    dist = euclidean_distances(combined_matrix, combined_matrix)

    # Get the index of the user input
    input_index = len(df) - 1
    distance_scores = list(enumerate(dist[input_index]))
    distance_scores = sorted(distance_scores, key=lambda x: x[1])

    # Get top 'jumlah' similar products
    top_indices = [i for i, _ in distance_scores[1:int(jumlah)+1]]

    # Prepare the recommendation output
    rekomendasi = df.iloc[top_indices][['name', 'brand', 'rating', 'price', 'sub-category']].copy()
    rekomendasi['euclidean_distance'] = [distance_scores[i][1] for i in range(1, int(jumlah)+1)]

    return rekomendasi.reset_index(drop=True), combined_matrix, tfidf_benefit, tfidf_ing, tfidf_subcat

```

```python
hasil, combined_matrix, tfidf_benefit, tfidf_ing, tfidf_subcat = get_recommendations(
    input_benefit="memberikan hidrasi yang tahan lama serta meningkatkan kecerahan kulit, membantu menutrisi, memperbaiki, dan menenangkan kulit yang sensitif atau kusam.",
    input_ingredients="WATER, BUTYLENE GLYCOL, GLYCERIN, 1,2-HEXANEDI-OL, TREHALOSE, NIACINAMIDE, BIFIDA FERMENT FILTRATE, PEG-12 DIMETHICONE CROSSPOLYMER, HYDROXYACETOPHENONE, AMMONIUM ACRYLOYLDI-METHYLTAURATE/BEHENETH-25 METHACRYLATE, CROSSPOLYMER, 2-0-ETHYL ASCORBIC ACID, ALPHA-ARBUTIN, ALLANTOIN, PROPYLENE GLYCOL, DISODIUM EDTA, PEG-40 HYDROGENATED CASTOR OIL, BACILLUS/SOYBEAN FERMENT EXTRACT, 3-0-ETHYL ASCORBIC ACID, CITRUS AURANTIUM DULCIS PEEL OIL, CITRUS RETICULATA PEEL EXTRACT, ASCORBIC ACID, ASCORBYL TETRAISOPALMITATE, ASCORBYL PALMITATE, SODIUM HYALURONATE, ACTINIDIA CHINENSIS FRUIT EXTRACT, FOLIC ACID",
    input_subcat="Toner",
    jumlah=5
)

```
